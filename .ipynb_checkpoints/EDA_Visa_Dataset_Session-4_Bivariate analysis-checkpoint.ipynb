```python
# Create a dataframe

#          Certified    Denied
#  Asia    11012        5849
#  Africa  397          154
```

```python
certified,denied=[],[]
for i in continents:
    con1=(visa_dataset['continent']==i)
    con2=(visa_dataset['case_status']=='Denied')
    con3=(visa_dataset['case_status']=='Certified')
    count_denied=len(visa_dataset[con1&con2])
    count_certified=len(visa_dataset[con1&con3])
    certified.append(count_certified)
    denied.append(count_denied)
```

```python
pd.DataFrame(zip(certified,denied),
            columns=['Certified','Denied'],
            index=visa_dataset['continent'].unique())
```

**Pearson-Correlation-Coeffiecint(r)**

```python
- 100  to 100

r=-1  to   1


-1  to 0 ======== > negative relation

-0.5 : both fetaures 50% negatively correlated
    
0 to 1 ========= > postively related
    
0.7: both features 70% postively correlated
    
0 : No relation/ independent each other
```

```python
r will vary  -1 to 1

-1 to 0 ====== negative

0 to 1 ======= post

=
```
