## METHOD 1

```python
cols_dict = dict(visa_dataset.dtypes)
cat_list,num_list=[],[] # declaring two list variables (categorical, and numerical)
for key, value in cols_dict.items():
    if value == 'O':
        cat_list.append(key)
    else:
        num_list.append(key)
        
cat_list, num_list

# whenever value == 'object'I'll append the keys in a list
# then the list becomes categorical columns

```

## LIST COMPREHENSION METHOD 

```python
cat_list = [key for key, value in cols_dict.items() if value == 'O']
num_list = [key for key, value in cols_dict.items() if value != 'O']
print("Categorical data: ===>",cat_list)
print("Numerical data: ===>",num_list)
```
